## 라이브러리 import

```python
import cv2
#matplotlib과 연동하기 위함
import matplotlib.pyplot as plt 
```

```python
#imread() : 이미지를 읽는 기능
# cv2.IMREAD_COLOR : 컬러 이미지를 읽는 설정

lenna = cv2.imread('./data/lenna.png', cv2.IMREAD_COLOR)

#BGR --> RGB , HSV 
lenna2 = cv2.cvtColor(lenna, cv2.COLOR_BGR2RGB)
```

```python
# 이미지를 출력하는 기능 (타이틀, 이미지)
cv2.imshow('lenna', lenna)

# 아무 키나 누르는 것을 대기
cv2.waitKey(0)

# 열린 창 모두 닫기
cv2.destroyAllWindows()
```

```python
#matplotlib과 이미지 연동

plt.imshow(lenna2)

# 눈금 생략
plt.xticks([])
plt.yticks([])
```

### 흑백이미지로 불러오기
- 컬러 이미지 : RGB 등의 색상으로 이루어진 이미지
- 그레이 이미지 : 0 -255 사이의 픽셀로 이루어진 이미지
- 흑백 이미지 (이진 이미지) : 0과 255 또는 0과 1의 픽셀로 이루어진 이미지

```python
lenna_gray = cv2.imread('./data/lenna.png', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(lenna_gray, cmap="gray")
plt.xticks([])
plt.yticks([])
```

```python
# 이진 이미지

#이미지 픽셀값을 127보다 큰 값은 255, 작은 값은 0으로 변환
_, thlmg = cv2.threshold(lenna_gray, 160, 255, cv2.THRESH_BINARY)

plt.imshow(thlmg, cmap="gray")
plt.xticks([])
plt.yticks([])
```
